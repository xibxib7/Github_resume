```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Top_Youtubers_Dataset.csv')
```

```python
df
```

```python
print("THIS IS : Explore the Data")
```

```python
print("Display the first few rows of the DataFrame")
df.head()
```

```python
print(" Get summary statistics")

df.describe()
```

```python
print(" Check the data types and missing values")

df.info()
```

```python
print("THIS IS : Analyzing the Data:")
```

```python
plt.figure(figsize=(10, 6))
```

```python
print("\nPlot the distribution of subscribers")
plt.figure(figsize=(10, 6))
plt.hist(df['Subscribers'], bins=15, color='skyblue', edgecolor='black')
plt.title("Distribution of Subscribers")
plt.xlabel("Subscribers(in millions)")
plt.ylabel("Frequency")
plt.show()
```

```python
print("\n top 10 channels by subscribe")

top10_subs = df.nlargest(10, 'Subscribers')
plt.figure(figsize=(12, 6))
plt.bar(top10_subs['Channel Name'], top10_subs['Subscribers'], color='lightgreen')
plt.title('Top 10 Channels by Subscribers')
plt.xlabel('Channel Name')
plt.ylabel('Subscribers (in milions)')
plt.xticks(rotation=45)
plt.show()
```

```python
print(" Subscribers vs Video Views:")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Subscribers'], df['Video Views'], color='coral')
plt.title('Subscribers vs Video Views')
plt.xlabel('Subscribers(in millions)')
plt.ylabel('Video Views(in bilions)')
plt.show()
```

```python
print("d. Average Subscribers by Category")
```

```python
avg_subs_by_category = df.groupby('Category')['Subscribers'].mean().sort_values()
plt.figure(figsize=(12, 6))
avg_subs_by_category.plot(kind='barh', color='lightblue')
plt.title('Average Subscribers by Category')
plt.xlabel('Average Subscribers(in milion)')
plt.ylabel('Category')
plt.show()
```

```python
print("a. Growth Over Time:")
print(" Calculate the age of each channel")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Category'], df['Channel Name'], color='purple')
plt.title('Channel Age vs Subscribers')
plt.xlabel('Age of Channel(years)')
plt.ylabel('Subscribe(in milion)')
plt.show()
```

```python
print(" Video Views per Subscriber:")
```

```python
print("\n Calculate video views per subscriber")
df['Views per Subscriber'] = df['Video Views'] / df['Subscribers']

plt.figure(figsize=(12, 6))
plt.bar(df['Channel Name'], df['Views per Subscriber'], color='red')
plt.title('Video Views per Subscriber for Each Channel')
plt.xlabel('Channel Name')
plt.ylabel('View per Subscriber')
plt.xticks(rotation=90)
plt.show()
```

```python
print("c. Correlation Analysis:")
```

```python
#Correlation matrix

correlation = df[['Subscribers', 'Video Views', 'Video Count','Views per Subscriber']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True)
```
